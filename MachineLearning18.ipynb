```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
df=load_iris()
print(dir(df))
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
df=load_iris()
print(dir(df))
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
df=pd.read_csv("Income.csv")
df
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
df[["Age","Income($)"]]=scaler.fit_transform(df[["Age","Income($)"]])
df
```

```python
model=KMeans(n_clusters=3)
y_prediction=model.fit_predict(df[["Age","Income($)"]])
y_prediction
```

```python
df["clusters"]=y_prediction
df
```

```python
cond1=df["clusters"]==0
cond2=df["clusters"]==1
cond3=df["clusters"]==2
plt.scatter(df[cond1]["Age"],df[cond1]["Income($)"],color="green")
plt.scatter(df[cond2]["Age"],df[cond2]["Income($)"],color="blue")
plt.scatter(df[cond3]["Age"],df[cond3]["Income($)"],color="yellow")
plt.scatter(model.cluster_centers_[:,0:1],model.cluster_centers_[:,1:2],label="centroid",color="orange",marker="*")
```

```python
mse=[]
krange=range(1,10)
for k in krange:
    model=KMeans(n_clusters=k)
    model.fit(df[["Age","Income($)"]])
    mse.append(model.inertia_)
plt.plot(krange,mse,marker="o")
```
